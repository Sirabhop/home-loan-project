# Mapping data

```python
import os
import pandas as pd
```

```python
file_names = os.listdir('data/')
csv_files = [f for f in file_names if f.endswith('.csv')]
```

```python
application_df = pd.read_csv("data/application_train.csv")

# List to store the percentage of matching records for each dataset
matching_percentages = []

# Iterate through CSV files (excluding "application.csv")
for file_name in csv_files:
    if file_name != "application.csv":
        # Load the current dataset
        dataset_df = pd.read_csv(os.path.join('data/', file_name))

        # Perform an inner join on 'SK_ID_CURR'
        merged_df = pd.merge(application_df, dataset_df, on='SK_ID_CURR', how='inner')

        # Calculate the percentage of matching records
        percentage = (len(merged_df) / len(application_df)) * 100
        matching_percentages.append((file_name, percentage))

# Print the results
for file_name, percentage in matching_percentages:
    print(f"{file_name}: {percentage:.2f}% of records match with the 'application' dataset")
```
